### Using the csv file since the geocoder is inconsistent

```python
import pandas as pd
import numpy as np
import requests
import csv
```

### Importing dataframe from part 1

```python
torontohood = pd.read_csv('torontoneighborhood.csv')
```

```python
torontohood.drop(columns='Unnamed: 0',inplace=True)
```

```python
torontohood
```

### downloading the coordinate csv file and opening it as pandas df

```python
url = 'http://cocl.us/Geospatial_data'
req = requests.get(url)
cont = req.content
csv_file=open('latlong.csv','wb')
csv_file.write(cont)
csv_file.close()
latlong = pd.read_csv('latlong.csv')
```

### sorting orders of both dataframes so it matches

```python
latlong.sort_values(by=['Postal Code'],inplace=True)
torontohood.sort_values(by=['PostalCode'],inplace=True)
torontohood.reset_index(drop=True,inplace=True)
latlong.reset_index(drop=True,inplace=True)
```

```python
latlong    
```

```python
torontohood
```

```python
torontohood['Latitude']=np.nan
torontohood['Longitude']=np.nan
torontohood
```

### adding latitude and longitude row by row to make sure postalcode matches

```python
for index,row in torontohood.iterrows():
    if row['PostalCode'] == latlong.iloc[index,0]:
        torontohood.iloc[index,3] = latlong.iloc[index,1]
        torontohood.iloc[index,4] = latlong.iloc[index,2]
    else:
        print('error')
        exit()
```
